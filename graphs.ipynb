Topic Distribution Graph Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import squarify

# Read data from CSV file
filename = 'Topic_Counts.csv'
data = pd.read_csv(filename)

# Calculate the total count for each topic
topic_counts = {}
for topic, count in zip(data['Topic'], data['count']):
    topic_counts[topic] = topic_counts.get(topic, 0) + count

# Prepare data for squarify
sizes = [topic_counts[topic] for topic in sorted(topic_counts.keys())]
labels = sorted(topic_counts.keys())

# Plotting
plt.figure(figsize=(10, 7))
squarify.plot(sizes=sizes, label=labels, alpha=0.7, text_kwargs={'fontsize': 10, 'verticalalignment': 'center'})

plt.axis('off')
plt.show()
```
